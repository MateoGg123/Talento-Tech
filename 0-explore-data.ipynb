```python
# Se ha cargado la variable "df" del URI: /workspaces/codespaces-jupyter/medium_articles.csv
import pandas as pd
df = pd.read_csv(r'/workspaces/codespaces-jupyter/medium_articles.csv')

# Filtrar filas según la columna: 'tags'
df = df[df['tags'].str.contains("technology", regex=False, na=False, case=False)]

# Filtrar filas según la columna: 'tags'
df = df[df['tags'].str.contains("technology", regex=False, na=False)]

# Derivar la columna 'derivedCol' de la columna: 'timestamp'
df.insert(5, 'derivedCol', df.apply(lambda row : "", axis=1))

# Derivar la columna 'derivedCol_1' de la columna: 'timestamp'
# Transformación basada en los siguientes ejemplos:
#    timestamp                             Salida
# 1: "2020-10-10 20:17:37.132000+00:00" => "2020-10-10"
df.insert(5, "derivedCol_1", df["timestamp"].str.split(" ").str[0])

# Eliminar columna: 'derivedCol'
df = df.drop(columns=['derivedCol'])

# Derivar la columna 'derivedCol' de la columna: 'derivedCol_1'
df.insert(6, 'derivedCol', df.apply(lambda row : "", axis=1))

# Eliminar columna: 'derivedCol'
df = df.drop(columns=['derivedCol'])

# Eliminar columna: 'derivedCol_1'
df = df.drop(columns=['derivedCol_1'])

# Derivar la columna 'derivedCol' de la columna: 'timestamp'
# Transformación basada en los siguientes ejemplos:
#    timestamp                             Salida
# 1: "2020-10-10 20:17:37.132000+00:00" => "2020-10-10"
df.insert(5, "derivedCol", df["timestamp"].str.split(" ").str[0])

# Cambia el tipo de columna por datetime64[ns] para la columna: 'derivedCol'
df = df.astype({'derivedCol': 'datetime64[ns]'})

# Filtrar filas según la columna: 'derivedCol'
from datetime import datetime
df = df[df['derivedCol'] > datetime.strptime('2021-01-01T00:31:47.183Z', '%Y-%m-%dT%H:%M:%S.%fZ')]

# Eliminar las columnas: 'title', 'url' y 4 otras columnas
df = df.drop(columns=['title', 'url', 'authors', 'timestamp', 'tags', 'derivedCol'])

# Longitud de texto calculada de la columna: 'text'
df.insert(1, 'text_len', df['text'].str.len())
df.head()
```
